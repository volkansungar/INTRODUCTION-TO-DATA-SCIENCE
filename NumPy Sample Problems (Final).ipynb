# Problem-1

It is requested to calculate the BMI (body mass index) values of football players with a height list given in inches and a weight list given in pounds. (To convert inches to meters, multiply by 0.0254, to convert pounds to kilograms, multiply by 0.453592)

```python
heights = [79.13385826771654, 72.83464566929134, 70.86614173228347, 73.22834645669292, 74.01574803149607, 75.59055118110236, 71.65354330708662, 72.04724409448819]
weights = [185.82544665690753, 178.57457803488597, 174.16532919451842, 189.59770013580487, 205.03007107709132, 191.80232455598863, 171.96070477433466, 153.5514559339671]
```

```python
import numpy as np
```

```python
#Create an array named heights_m containing the heights in meters, using the values in the heights list
heights_m =np.array(heights)*0.0254
print(heights_m)
heights_m
```

```python
#Create an array named weights_kg containing the weights in kilograms, using the values in the weights list
weights_kg =np.array(weights)*0.453592
weights_kg
```

```python
#Print the data from the 3rd index to the 6th index (inclusive) of the weights_kg array.
print(weights_kg[3:7])
```

```python
#Calculate BMI of football players
bmi=weights_kg/heights_m**2
bmi
```

```python
#Create an array named talls containing heights greater than 1.90 meters
talls=heights_m[heights_m>1.90]
print(talls)
```

```python
#Create an array that includes "normal" for those with a BMI below 25 and "overweight" for others. 
np.where(bmi<25,"normal","overweight")
```

# Problem-3

Create 5 randomly generated cylinders with radius and height ranging from 5 to 25 cm.

a) Calculate the volumes of all cylinders.

b) Find the total volume of all cylinders.

c) Find the volume and index of the cylinder with the largest volume.

d) Find the mean, median and standard deviation values of the volumes of the cylinders.

## First, create an array containing 10 random integers with a lower bound of 5 and an upper bound of 25.

```python
rng = np.random.default_rng(seed=12345)
integers = rng.integers(5,25,10)
integers
```

## Then set the array to two dimensions for 5 cylinders and create a cylinder array.

```python
cylinders =integers.reshape(5,2)
cylinders
```

## \begin{equation*} 
\text{cylinder_volume=} 
\pi r^2h
\ 
\end{equation*}

## a-Calculate the volumes of all cylinders.

```python
volumes = np.pi*cylinders[:,0]**2*cylinders[:,1]
volumes
```

## b-Find the total volume of all cylinders.

```python
total_volume = np.sum(volumes)
#total_volume = volumes.sum()
total_volume
```

## The total volume can also be found in an alternative way without calculating the volumes of individual cylinders (dot product of radii and heights).

```python
total_volume_2 = np.pi*np.dot(cylinders[:,0]**2, cylinders[:,1])
total_volume_2
```

## c-Find the volume and index of the cylinder with the largest volume.

```python
max_volume = volumes.max()
max_volume_index = np.argmax(volumes)
print(max_volume,max_volume_index)
```

## d-Find the mean, median and standard deviation values of the volumes of the cylinders.

```python
volume_mean = np.mean(volumes)
volume_median = np.median(volumes)
volume_std = np.std(volumes)
print(volume_mean,volume_median,volume_std)
```
